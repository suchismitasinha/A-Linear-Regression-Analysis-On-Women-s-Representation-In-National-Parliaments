```
#Import the necessary libaries
from pyspark.ml.regression import LinearRegression
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.feature import StandardScaler
from pyspark.ml import Pipeline
from pyspark.sql.functions import *
```

```
#Set a variable with the path
file_name = "/FileStore/tables/Seats.csv"
```

```
#Load the file into a dataframe and perform simple exploratory data analysis
data = spark.read.options(header='true', inferschema='true', delimiter=',').csv(file_name)
data.cache()
display(data.describe())
```

```
# Convert the seats_csv SQL table into a DataFrame named dataset. We put all features into a vector and since we are dealing with numerical data, we scale those features
# TODO: Replace <FILL_IN> with the appropriate code
from pyspark.ml.feature import VectorAssembler

#datasetDF = <FILL_IN>
datasetDF = sqlContext.table('seats_csv')

vectorizer = VectorAssembler()
#vectorizer = VectorAssembler(inputCols=["1990", "1991", "1992", "1993", "1994", "1995", "1996", "1997", "1998", "1999", "2000", "2001", "2002", "2003", "2004", "2005", "2006",	"2007",	"2008",	"2009",	"2010",	"2011",	"2012",	"2013",	"2014",	"2015",	"2016",	"2017", "2018"], outputCol="features")

#vectorizer.setInputCols(<FILL_IN>)
vectorizer.setInputCols(["1990", "1991", "1992", "1993", "1994", "1995", "1996", "1997", "1998", "1999", "2000", "2001", "2002", "2003", "2004", "2005", "2006", "2007","2008",	"2009",	"2010",	"2011",	"2012",	"2013",	"2014",	"2015",	"2016",	"2017","2018"])
                         
#vectorizer.setOutputCol(<FILL_IN>)
vectorizer.setOutputCol("features")
#vectorizer = VectorAssembler(inputCols=["1990", "1991", "1992", "1993", "1994", "1995", "1996", "1997", "1998", "1999", "2000", "2001", "2002", "2003", "2004", "2005", "2006", "2007",	"2008",	"2009",	"2010",	"2011",	"2012",	"2013",	"2014",	"2015",	"2016",	"2017", "2018"], outputCol="features")
```

```
#The dataset is split into the training and the test set where 20% of the data is alloted for the test data and the remaining 80% as training data
#(split20DF, split80DF) = datasetDF.<FILL_IN>
(split20DF, split80DF) = data.randomSplit([0.2, 0.8])

# Let's cache these datasets for performance
#testSetDF = <FILL_IN>
testSetDF = split20DF.cache()

#trainingSetDF = <FILL_IN>
trainingSetDF = split80DF.cache()

print(trainingSetDF)
```

```
#Importing necesary libaries for our Linear Regression model
from pyspark.ml.regression import LinearRegression
from pyspark.ml.regression import LinearRegressionModel
from pyspark.ml import Pipeline

# Let's initialize our linear regression learner
lr = LinearRegression()

# We use explain params to dump the parameters we can use
print(lr.explainParams())
```

```
# We build the machine learning pipeline here
# Now we set the parameters for the method
lr.setPredictionCol("Predicted")\
  .setLabelCol("2018")\
  .setMaxIter(200)\
  .setRegParam(0.1)

# We will use the new spark.ml pipeline API
lrPipeline = Pipeline()

lrPipeline.setStages([vectorizer, lr])

# We first train on the entire dataset to see what we get
lrModel = lrPipeline.fit(trainingSetDF)
```

```
#Make the predictions based on our label and features
predictionsAndLabelsDF = lrModel.transform(testSetDF).select("country","2018","features", "Predicted")
display(predictionsAndLabelsDF)
```

```
# We compute an evaluation metric for our test dataset
from pyspark.ml.evaluation import RegressionEvaluator

# We create an RMSE evaluator using the label and predicted columns
regEval = RegressionEvaluator(predictionCol="Predicted", labelCol="2018", metricName="rmse")

# Run the evaluator on our DataFrame
rmse = regEval.evaluate(predictionsAndLabelsDF)

print("Root Mean Squared Error: %.2f" % rmse)
```

```
#Calculation of the following metrics
# Mean Square Error
mse = regEval.evaluate(predictionsAndLabelsDF, {regEval.metricName: "mse"})
print("Mean Square Error: %.3f" % mse)

# Mean Absolute Error
mae = regEval.evaluate(predictionsAndLabelsDF, {regEval.metricName: "mae"})
print("Mean Absolute Error: %.3f" % mae)

#r2 - coefficient of determination
r2 = regEval.evaluate(predictionsAndLabelsDF, {regEval.metricName: "r2"})
print("r2 - coefficient of determination: {0:.2f}".format(r2))
```
